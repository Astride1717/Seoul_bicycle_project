```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

minwon = []
for i in range(4159, 60384):
    url = 'https://www.bikeseoul.com/customer/opinionBoard/opinionBoardView.do?usrQnaSeq='+str(i)
    if str(i) in url: 
        html = urlopen(url)
        soup_tmp = BeautifulSoup(html, "html.parser")
        title = soup_tmp.find('p').get_text().strip() #제목
        content = soup_tmp.find('dd').get_text().strip() #내용
        minwon.append(title)
        minwon.append(content)
```
